```python

# importing pandas as pd
import pandas as pd

# importing re for regular expressions
import re

# Creating the Series
sr = pd.Series(['Mike', 'Alessa', 'Nick', 'Kim', 'Britney'])

# Creating the index
idx = ['Name 1', 'Name 2', 'Name 3', 'Name 4', 'Name 5']

# set the index
sr.index = idx

# Print the series
print(sr)
```

```python
# find if there is a substring such that it has
# the letter 'i' followed by any small alphabet.
result = sr.str.contains(pat = 'i[a-z]', regex = True)
#result = sr.str.encode(encoding = 'raw_unicode_escape')
# print the result
print(result)
```

```python

# match either 'Tokyo' or 'Paris'
result = sr.str.match(pat = '(Mike)|(Kim)')

# print the result
print(result)
```
