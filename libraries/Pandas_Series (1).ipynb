# **Indexing, Slicing, and Selection in Pandas Series**
Indexing, slicing, and selection are fundamental operations that allow you to access and manipulate data within a pandas Series. These operations are crucial for data engineering and data science tasks where you might need to filter, manipulate, or access specific elements within a Series.

# **Accessing Elements by Index**
You can access elements in a Series by their index, which can be either an integer (default) or a custom index (if provided).

````python
# Create a Series with default integer index
s = pd.Series([10, 20, 30, 40])

# Accessing elements by integer index
print(s[0])  # First element
print(s[2])  # Third element
````

````python
# Create a Series with custom index
s = pd.Series([100, 200, 300], index=['a', 'b', 'c'])

# Accessing elements by custom index
print(s['a':'c'])  # Range of Elements from index 'a' to 'c'
print(s['b'])  # Element with index 'b'
````

# **Slicing in pandas Series**
Basic Slicing (Integer Indexing)
You can slice a Series just like you would with a Python list, using the :
    operator. This allows you to get a range of values or subsets.

````python
# Create a Series with default index
s = pd.Series([10, 20, 30, 40, 50])

# Slicing the Series
print(s[1:])  # Elements from index 1 to the end
````

````python
# Create a Series with custom index
s = pd.Series([100, 200, 300, 400], index=['a', 'b', 'c', 'd'])

# Slicing the Series with custom index
print(s['b':'d'])  # Elements from index 'b' to 'd' (inclusive)
````

 # 2- Selection Using .at[] and .iat[]
at[]: Used for getting or setting a single value by index label (faster for scalar access).

iat[]: Used for getting or setting a single value by integer position (faster for scalar access).

````python
# Create a Series with custom index
s = pd.Series([100, 200, 300], index=['a', 'b', 'c'])

# Using .at[] for scalar access (label-based)
print(s.at['b'])  # Access value at index 'b'

# Using .iat[] for scalar access (integer-based)
print(s.iat[1])  # Access value at position 1
````

 #  3- Selection by Index Using .loc[] and .iloc[]
loc[]: Used for label-based indexing (custom index).

iloc[]: Used for integer-based indexing (default integer index).

#.iloc[] (Integer-based Indexing)
This is used when you want to select elements by their integer position, regardless of the index label

````python
# Create a Series with custom index
s = pd.Series([100, 200, 300, 400], index=['a', 'b', 'c', 'd'])

# Using .iloc[] to access elements by integer position
print(s.iloc[1])  # Access second element (position 1)s
print(s.iloc[0:3])  # Slice from position 0 to 2 (3 is excluded)
````

#.loc[] (Label-based Indexing)
This allows you to select values based on index labels. The index labels can be strings or any other data type you've specified during initialization.
# **Possible types of things you can put inside the brackets:**
1. Single label	------>**s.loc['a']**------>	Returns one element

2. List of labels------>	**s.loc[['a', 'b']]**------>	Returns multiple elements


3. Label slice	------>   **s.loc['a':'c']**	------>       Returns range (inclusive)

4. Boolean mask ------>**s.loc[s > 10]**	------>Filters values

5. Callable/function------>	**s.loc[lambda x: x > 10]**------>	Same as boolean mask

6. Assignment ------>	**s.loc['a'] = 999**	------>Updates values in the original Series



# Boolean mask
A boolean mask is a list (or Series) of True and False values — one for each element in the Series.

True = Keep this element

False = Ignore this element

```
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
Index     =>   ['a', 'b', 'c']
Values    =>   [10,   20,   30]
Mask      =>   [ T,    F,    T ] → Keep 'a' and 'c'
```
**You can build this mask dynamically**  
**1. conditions:**
```
s = pd.Series([10, 20, 30, 40])
mask = s > 15  # Gives: [False, True, True, True]
s.loc[mask]
#Output: 20, 30, 4
```

**2. Using Lambda Function**
```
s.loc[lambda x: x > 20]

x is the Series s
x > 20 creates the boolean mask: [False, True, True, False]
.loc[] filters based on that mask
```

````python
s = pd.Series([10, 20, 30, 40, 50])
mask = (s > 20) & (s < 50)
print(s.loc[mask])
print('--------------------------')
print(s.loc[lambda x: (x > 20) & (x < 50)])
````

````python
# 7. Modify with Conditions + Positions is very important
s = pd.Series([10, 20, 30, 40, 50])

# Get indices where s > 20
indices = s[s > 20].index[:2]  # First 2 values greater than 20

# Modify them
s.loc[indices] = 777
print(s)
````

#Modifying the first N values that meet a condition

````python
import pandas as pd
s = pd.Series([10, 20, 30, 40, 50])

# Step 1: Find positions where condition is True
mask = s > 20

# Step 2: Get the first 3 indices where condition is True
indices_to_change = s[mask].index[:3]         # s[mask][:3]=100 isn't correct for your task

# Step 3: Modify those indices directly in the original Series
s.loc[indices_to_change] = 100

print(s)
````

#**6. Index Operations**

#**reindex**()  

changes the index of a Series to a new one that you provide.
If the new index contains values not in the original, pandas will fill those spots with NaN.


````python
# You have time series data representing daily sales for a retail store.
# The data has some missing dates, and you need to reindex the data to ensure that it's continuous
#  — i.e., fill in the missing dates.
#****************************************************************************************************
#Step 1: Creating the Data
# Sample time series data (with missing dates)
data = {
    '2025-01-01': 100,
    '2025-01-03': 120,
    '2025-01-04': 130,
    '2025-01-07': 110,
    '2025-01-08': 150
}

# Create the Series with a datetime index
s = pd.Series(data=data.values(), index=pd.to_datetime(list(data.keys())))

print("Original Sales Data:")
print(s)
#***********************************************************************
# Step 2: Reindexing with Missing Dates , The problem here is that we have missing dates (2025-01-02, 2025-01-05, 2025-01-06 are missing).
#  We want to reindex the Series so that it includes all dates
#   from 2025-01-01 to 2025-01-08, filling in the missing dates.
# Create a date range from the start to end date (continuous daily dates)
new_index = pd.date_range('2025-01-01', '2025-01-08')

# Reindex the Series to fill in the missing dates
reindexed_s = s.reindex(new_index)

print("\nReindexed Sales Data with Missing Dates (NaN filled):")
print(reindexed_s)
````

# **String Operations on Series**
When working with a Series of strings (text), pandas provides a .str accessor, which gives you vectorized string functions (like you get with Python strings, but fast and applicable to the whole column).

image.png

````python
s = pd.Series(['apple', 'banana', 'orange', 'banana', None])
s.dropna(inplace=True)
s
# Replace 'banana' with 'mango'
s.replace('banana','mango',inplace=True)

# find index of certain elements
s.index[s.str.contains('mango')]

# Convert to uppercase and filter entries that contain 'E'
s_upper = s.str.upper()[s.str.contains('o')]
s_upper
````

````python
import pandas as pd

# Sample Series
s = pd.Series([
    '  DataScience ',     # leading/trailing whitespace
    'AI & ML',            # uppercase, ampersand
    'data@123',           # symbols and numbers
    'Deep Learning',      # two words
    '',                   # empty string
    None                  # missing value
])

# 1. Change Case
print("Lowercase:\n", s.str.lower())
print("\nUppercase:\n", s.str.upper())
print("\nTitle Case:\n", s.str.title())

# 2. Strip Spaces
print("\nStripped:\n", s.str.strip())

# 3. String Length
print("\nString Lengths:\n", s.str.len())

# 4. Contains (substring check)
print("\nContains 'data':\n", s.str.contains('data', case=False, na=False))

# 5. Startswith / Endswith
print("\nStarts with 'D':\n", s.str.startswith('D', na=False))
print("\nEnds with 'g':\n", s.str.endswith('g', na=False))

# 6. Replace substring
print("\nReplace '@' with '_':\n", s.str.replace('@', '_', regex=False))

# 7. Split string into list
print("\nSplit on space:\n", s.str.split())

# 8. Join strings (only works if elements are lists)
split_series = s.str.split()
print("\nJoin with '-':\n", split_series.dropna().apply(lambda x: '-'.join(x)))

# 9. Extract regex (first number group)
print("\nExtract numbers:\n", s.str.extract(r'(\d+)'))

# 10. Find all regex matches
print("\nFind all letters:\n", s.str.findall(r'[a-zA-Z]+'))

# 11. Access individual character
print("\nFirst character:\n", s.str[0])

# 12. Check empty and missing
print("\nIs Empty String:\n", s == '')
print("\nIs NaN:\n", s.isna())
````

#**Data Cleaning and Preprocessing (filling NaN, filtering)**

It’s the process of:

Handling missing or invalid data

Correcting data types

Filtering out outliers or irrelevant values

Making data ready for analysis or machine learning


Missing data in a Series is usually represented with None or NaN (Not a Number)

NaN (Not a Number) represents missing or undefined values. In pandas Series, it can come from:

Manual input

Incomplete data sources (e.g., CSV)

Operations that fail (e.g., divide by zero)

# Detecting Missing Data

````python
import pandas as pd
import numpy as np

s = pd.Series([1, np.nan, 3, None, 5])

# Boolean Series showing NaNs
print(s.isna())

# Count missing values
print("Count of NaNs:", s.isna().sum())

# Percentage of NaNs
nan_percentage = s.isna().sum() / len(s) * 100
print(f"NaN Percentage: {nan_percentage :.0f} %")
````

# Dropping Missing Values
Use this when NaNs are sparse and removing them won’t affect the analysis

````python
s = pd.Series([1, np.nan, 3, None, 5])
cleaned = s.dropna()
print(cleaned,)
````

#  Handling Missing Values (Filling NaNs)
1- Fill with a Fixed Value

2- Fill with Mean/Median/Mode (Statistical Imputation)

3- Forward/Backward Fill "Useful for time series or ordered data."



````python
s = pd.Series([1, np.nan, 3, None, 5],index=['a', 'b', 'c','d','e'])
print(f'the percentage of null values in the series is {s.isna().sum()/len(s)*100}%')
s_filled = s.fillna(33)
s_filled
````

````python
#Fill with Mean/Median/Mode (Statistical Imputation)
s = pd.Series([1, np.nan, 3, None, 5],index=['a', 'b', 'c','d','e'])
print(s.fillna(s.mean()))

s = pd.Series([1, np.nan, 3, None, 5],index=['a', 'b', 'c','d','e']) #1,3,5 ->3
print(s.fillna(s.median()))

s = pd.Series([1, np.nan,2,3,5, None, 5],index=['a', 'b', 'c','d','e','f','g']) #mode 5
print(s.fillna(s.mode()[0]))
````

````python
#Forward/Backward Fill useful for time series or ordered data.
s = pd.Series([1, np.nan, 3, None, 5],index=['a', 'b', 'c','d','e'])
s_filled  = s.fillna(method='ffill')  # Forward fill
print(s_filled)

s = pd.Series([1, np.nan, 3, None, 5],index=['a', 'b', 'c','d','e'])
s.fillna(method='bfill')  # Backward fill
print(s_filled)
````
